# Here are the business questions

```python
#Q1)Total number of transactions has been done state wise
tot_trans_state_wise=stores.groupby(['state'])[['transactions']].sum()
tot_trans_state_wise
```

```python
#plot the bar graph
tot_trans_state_wise['transactions'].plot.bar(figsize=(16,8))
```

```python
#Q2)which state got the heightest transaction
tot_trans_state_wise.loc[tot_trans_state_wise['transactions'].idxmax()]
```

```python
#which state got the minimum transaction
tot_trans_state_wise.loc[tot_trans_state_wise['transactions'].idxmin()].plot.bar(figsize=(12,8))
```

```python
#merge the dataset with oil dataset
oil['date']=pd.to_datetime(oil['date'])
stores=stores.merge(oil,on='date',how='left')
del oil
```

```python
#replace the NaN values with mean values of oil price
mean_oil_price=(np.mean(stores.dcoilwtico))
stores.fillna(mean_oil_price,inplace=True)
stores
```

```python
#Q4)what are the average oil price in every year and in whih year oil prices were increased too much on an average
mean_oil_price=stores.groupby(['year'])[['dcoilwtico']].mean()
mean_oil_price
```

```python
mean_oil_price['dcoilwtico'].plot.bar(figsize=(12,8))
```

```python
#Q5)which family has the lowest item numbers
low_item_total=items.groupby(['family'])[['item_nbr']].count()
low_item_total
```

```python
#plot the bar chart
low_item_total.sort_values('item_nbr',ascending=True).plot.bar(figsize=(12,8))
```
